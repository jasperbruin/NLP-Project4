```python
import json

def load_data(filepath):
    """ Load JSON data from a file. """
    with open(filepath, 'r') as file:
        data = json.load(file)
    return data

def extract_claims(data):
    """ Extract claims from the loaded JSON data by isolating the text after '[Claim]:' and before '[Evidences]:'. """
    claims = []
    for item in data:
        claim_evidence_string = item['claim_evidence_string']
        claim_start_idx = claim_evidence_string.find("[Claim]:") + len("[Claim]:")
        evidences_start_idx = claim_evidence_string.find("[Evidences]:")

        if evidences_start_idx != -1:
            claim = claim_evidence_string[claim_start_idx:evidences_start_idx].strip()
        else:
            claim = claim_evidence_string[claim_start_idx:].strip()

        if claim.startswith('"') and claim.endswith('"'):
            claim = claim[1:-1]
        claims.append(claim)
    return claims

def find_common_claims(file1, file2):
    """ Find common claims between two files. """
    data1 = load_data(file1)
    data2 = load_data(file2)

    claims1 = set(extract_claims(data1))
    claims2 = set(extract_claims(data2))

    return claims1.intersection(claims2)

# File paths (update these paths to your actual file locations)
file1 = '/content/drive/MyDrive/TUDelft/NLP Group Project/data/custom_claim_decomposition/custom_decomp_test.json'
file2 = '/content/drive/MyDrive/TUDelft/NLP Group Project/data/programfc/programfc_test.json'

# Execute the function to find common claims
common_claims = find_common_claims(file1, file2)

# Print the common claims
for claim in common_claims:
    print(claim)
```
